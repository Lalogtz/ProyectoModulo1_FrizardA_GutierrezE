**Objetivo general**

Realizar una proyección de la cantidad de personas en desempleo en Estados Unidos 

**Objetivos específicos**
+ Obtener la distribución de probabilidad que describa la población que se encuentra en desempleo
+ Obtener la distribución de probabilidad que describa la población que se encuentra empleada

### 1.3 Definición del problema.
Buscamos calcular las distribuciones de probabilidad de las variables de interés mencionadas en los objetivos específicos, a partir de los datos de desempleo en Estados Unidos con los que contamos.
